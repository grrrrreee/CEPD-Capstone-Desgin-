```python
import keras
```

```python
from keras.layers import Input, Dense, LSTM
```

```python
from keras.models import Model
```

```python
def build_model(input_shape):
    x = Input(shape=input_shape)

    h = LSTM(128, return_sequences=True, activation='tanh', kernel_initializer='he_normal')(x)
    h = LSTM(128, return_sequences=True, activation='tanh', kernel_initializer='he_normal')(h)
    h = LSTM(128, activation='tanh', kernel_initializer='he_normal')(h)

    y = Dense(1, activation='linear')(h)
    return Model(inputs=x, outputs=y)


```

```python
model = build_model((15, 4))
```

```python
model.summary()
```

```python
from keras.optimizers import Adam
```

```python
model.compile(loss='mse', optimizer=Adam(), metrics=['acc', 'mae'])
```
